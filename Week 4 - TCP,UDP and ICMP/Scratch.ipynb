```python
# CODE FOR SERVER

import socket

def start_server(host, port):
    #Create and start the UDP server.
    server_socket = socket.socket(socket.AF_INET, socket.SOCK_DGRAM)  # Create UDP socket
    server_socket.bind((host, port))  # Bind the server to the specified IP and port
    print(f"Server started on {host}:{port}")

    while True:
        try:
            # Receive message from the client
            message, client_address = server_socket.recvfrom(1024)  # buffer size is 1024 bytes
            print(f"Received message from {client_address}: {message.decode()}")

            # Send acknowledgment to the client
            ack_message = "Message received successfully!"
            server_socket.sendto(ack_message.encode(), client_address)

        except Exception as e:
            print(f"Error: {e}")
            break  # Exit the loop if an error occurs

    server_socket.close()

if __name__ == "__main__":
    host = "0.0.0.0"  # Listen on all available interfaces
    port = 8080
    start_server(host, port)
```

```python
# CODE FOR CLIENT

import socket
import sys

def send_message(server_ip, server_port, message):
    client_socket = socket.socket(socket.AF_INET, socket.SOCK_DGRAM)
    server_address = (server_ip, server_port)

    try:
        client_socket.sendto(message.encode(), server_address)
        print(f"Message sent to {server_ip}:{server_port}")

        ack, _ = client_socket.recvfrom(1024)
        print(f"Server acknowledgment: {ack.decode()}")

    except Exception as e:
        print(f"Error: {e}")

    finally:
        client_socket.close()

if __name__ == "__main__":
    if len(sys.argv) != 3:
        print("Usage: python udp_client.py <Server_IP> <Port>")
        sys.exit(1)

    server_ip = sys.argv[1]
    server_port = int(sys.argv[2])

    message = input("Enter the message to send: ")
    send_message(server_ip, server_port, message)
```
